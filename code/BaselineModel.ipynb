```python
import numpy as np
import pandas as pd
```

```python
X_train = np.load('Data/Train/X.npy')
X_prime_train = np.load('Data/Train/X_prime.npy')
feature_info_train = np.load('Data/Train/feature_information.npy')
```

```python
# trainDataTrue = pd.DataFrame(data=X_train)
trainData = pd.DataFrame(data=X_prime_train)
```

```python
trainData.head()
```

```python
X_test = np.load('Data/Test/X.npy')
X_prime_test = np.load('Data/Test/X_prime.npy')
feature_info_test = np.load('Data/Test/feature_information.npy')
```

```python
# testDataTrue = pd.DataFrame(data=X_test)
testData = pd.DataFrame(data=X_prime_test)
```

```python
print("Train shape:",X_train.shape)
print("Test shape:",X_test.shape)
```

```python
categoricalFeatures = [[i for i in range(7,14)],
                 [i for i in range(14,30)],
                 [i for i in range(30,37)],
                 [i for i in range(37,51)],
                 [i for i in range(51,57)],
                 [i for i in range(57,62)],
                 [i for i in range(62,64)],
                 [i for i in range(64,105)]
                 ]
print(categoricalFeatures)
```

```python
isFeatureMissingTrain = pd.DataFrame(data = feature_info_train)
isFeatureMissingTrain.head()
```

```python
isFeatureMissingTest = pd.DataFrame(data = feature_info_test)
isFeatureMissingTest.head()
```

```python
class BaselineModel:
    
    def __init__(self, trainData, isFeatureMissingTrain, testData, isFeatureMissingTest, categoricalFeatures):
        self.features = isFeatureMissingTrain.columns
        self.trainData = trainData
        self.testData = testData
        self.categoricalFeatures = categoricalFeatures
        self.isCategorical = [i for _l in self.categoricalFeatures for i in _l]
        self.startIndex = [l[0] for l in self.categoricalFeatures]
        self.endIndex = [l[-1] for l in self.categoricalFeatures]
        
        # Missing value as Nan
        self.trainData = self.trainData.where(isFeatureMissingTrain > 0)
        self.testData = self.testData.where(isFeatureMissingTest > 0)
        
    def calculateSampleAverage(self):
        
        # Dictionary to store mean/mode of features
        self.meanValues = {}
        
        for col in self.features:
            # Categorical feature 
            if col in self.startIndex:
                start = col
                i = self.startIndex.index(col)
                end = self.endIndex[i]
                df = self.trainData.iloc[:,start:end+1]
                # Determine mode class
                colSum = np.array([df[c].sum() for c in df.columns])
#                 print("colSum",colSum)
                maxClass = np.argmax(colSum)
#                 print("maxClass:", maxClass)
                for i in range(start, end+1):
                    self.meanValues[i] = 0
                self.meanValues[maxClass+start] = 1
                
            elif col not in self.isCategorical:
                # Mean of numerical feature
                mean = self.trainData[col].mean()
                self.meanValues[col] = mean
            else:
                continue
        print(self.meanValues)
    
    def fillMissingValues(self):
        self.trainData.fillna(value = self.meanValues, inplace = True)
        self.testData.fillna(value = self.meanValues, inplace= True)
        
    # Root Mean square Error 
    def calculateError(self, trainTrueData, testTrueData):
        trainPredicted = self.trainData.to_numpy()
        testPredicted = self.testData.to_numpy()
        
        m, k = trainPredicted.shape
        m_test, k_test = testPredicted.shape
        
        diffTrain = (trainTrueData - trainPredicted)**2
        diffTest = (testTrueData - testPredicted)**2
        
        sumTrain = np.sum(diffTrain, axis = 1)
        sumTest = np.sum(diffTest, axis = 1)
        
        errTrain = np.mean(np.sqrt(sumTrain))
        errTest = np.mean(np.sqrt(sumTest))
        
        return errTrain, errTest
```

```python
baseline_model = BaselineModel(trainData, isFeatureMissingTrain, testData, isFeatureMissingTest, categoricalFeatures)
baseline_model.calculateSampleAverage()
baseline_model.fillMissingValues()
trainError, testError = baseline_model.calculateError(X_train, X_test)
```

```python
print("Training Error:", trainError)
print("Testing Error:", testError)
```
